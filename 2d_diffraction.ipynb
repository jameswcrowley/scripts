```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import image
import scipy.fftpack
```

###  Diffraction as a Fourier Transform of Apeture
*Predicting diffraction patterns from apeture shape*

```python
# creating an apeture 

apeture = np.ones((5001, 5001))
apeture[100:4900, 2495:2505] = 0

# finding pattern on screen as projection of fourier transform of apeture on a screen
pattern = scipy.fft.fftn(apeture)

plt.figure(figsize = [20, 8])
plt.subplot(1,2,1)
plt.title('apeture shape')
plt.imshow(apeture, cmap = 'gray'); plt.colorbar()

plt.subplot(1,2,2)
plt.title('pattern on screen')
plt.imshow(np.log(np.abs(scipy.fft.fftshift(pattern))), cmap = 'gray'); plt.colorbar()
```
